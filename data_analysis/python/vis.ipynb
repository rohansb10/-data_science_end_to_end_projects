```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data['Order_Quantity'], bins=30, kde=True)
plt.title('Distribution of Order Quantity')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Sales', data=data)
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Customer_Segment', y='Profit', data=data)
plt.title('Profit by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Order_Priority', data=data)
plt.title('Order Priority Distribution')
plt.xlabel('Order Priority')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Product_Category', y='Sales', data=data)
plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='Product_Sub-Category', y='Profit', data=data)
plt.title('Profit by Product Sub-Category')
plt.xlabel('Product Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Ship_Mode', data=data)
plt.title('Ship Mode Distribution')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=data)
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
top_10_products = data.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_products.index, y=top_10_products.values)
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.show()
```
